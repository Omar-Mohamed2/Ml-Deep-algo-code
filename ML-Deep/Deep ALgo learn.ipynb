functional api

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd

# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Reshape and normalize the input data
x_train = x_train.reshape(-1, 28 * 28) / 255.0
x_test = x_test.reshape(-1, 28 * 28) / 255.0

# One-hot encode the labels
y_train = tf.keras.utils.to_categorical(y_train, 10)
y_test = tf.keras.utils.to_categorical(y_test, 10)

# Print shapes after reshaping and one-hot encoding
print("Shapes after reshaping and one-hot encoding:")
print("x_train.shape:", x_train.shape)
print("y_train.shape:", y_train.shape)

# Display an image from the training set
plt.imshow(x_train[1].reshape(28, 28), cmap="gray")
plt.show()

# Define the model
def model(x, units=10, drop_rate=0.2):
    tf.keras.layers.Dropout(0.1)(x)
    layer1 = tf.keras.layers.Dense(units=units, activation=tf.nn.leaky_relu, name="layer1")(x)
    tf.keras.layers.Dropout(drop_rate)(layer1)
    concat_layer = tf.keras.layers.concatenate([x, layer1])
    return concat_layer

input_layer = tf.keras.layers.Input(shape=(784,), name="input_layer")
m1 = model(input_layer, 9, 0.1)
output_layer = tf.keras.layers.Dense(units=10, activation=tf.nn.softmax, name="output")(m1)

# Create a model
model = tf.keras.models.Model(inputs=input_layer, outputs=output_layer)

# Print model summary and plot the model
model.summary()
tf.keras.utils.plot_model(model, show_shapes=True, show_layer_names=True)

# Compile and train the model
model.compile(optimizer=tf.optimizers.Adam(), loss=tf.losses.CategoricalCrossentropy(), metrics=[tf.metrics.CategoricalAccuracy()])
model.fit(x=x_train, y=y_train, epochs=5, batch_size=30, validation_split=0.1)
h = model.history.history

# Plot training history
df = pd.DataFrame(h)
df.plot(figsize=(10, 8))
plt.show()
```

sub class api

```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

x_train = x_train.reshape(-1, 28 * 28)
x_test = x_test.reshape(-1, 28 * 28)
x_train, x_test = x_train / 255, x_test / 255

y_train, y_test = tf.one_hot(y_train, 10), tf.one_hot(y_test, 10)


class Neural(tf.keras.models.Model):
    def __init__(self, units, active_fun, drop_rate, **kwargs):
        super(Neural, self).__init__(**kwargs)
        self.units = units
        self.active_fun = active_fun
        self.drop_rate = drop_rate
        self.layer1 = tf.keras.layers.Dense(units=self.units, activation=self.active_fun)
        self.drop1 = tf.keras.layers.Dropout(self.drop_rate)
        self.layer2 = tf.keras.layers.Dense(units=self.units, activation=self.active_fun)
        self.drop2 = tf.keras.layers.Dropout(self.drop_rate)
        self.layer3 = tf.keras.layers.Dense(units=self.units, activation=self.active_fun)
        self.drop3 = tf.keras.layers.Dropout(self.drop_rate)
        self.out = tf.keras.layers.Dense(units=10, activation=tf.nn.softmax)

    def call(self, inputs):
        x = self.layer1(inputs)
        x = self.drop1(x)
        x = self.layer2(x)
        x = self.drop2(x)
        x = self.layer3(x)
        x = self.drop3(x)
        concat = tf.keras.layers.concatenate([x, inputs])
        out = self.out(concat)
        return out

    def build_graph(self):
        x = tf.keras.layers.Input(shape=(784,))
        return tf.keras.models.Model(inputs=[x], outputs=self.call(x))


# Create an instance of the Neural class
neural_model = Neural(units=128, active_fun='relu', drop_rate=0.3)
model = neural_model.build_graph()
model.compile(optimizer=tf.optimizers.Adam(), loss=tf.losses.CategoricalCrossentropy(), metrics=[tf.metrics.CategoricalAccuracy()])
model.fit(x=x_train, y=y_train, epochs=5, batch_size=30, validation_split=0.1)
model.summary()
tf.keras.utils.plot_model(model=model)
h = model.history.history
df = pd.DataFrame(h)
df.plot(figsize=(10, 8))
plt.show()

```
