linear regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets, linear_model, preprocessing, metrics, model_selection

# Load the iris dataset
x, y = datasets.load_iris(return_X_y=True)

# Standardize the input features
x_normalize = preprocessing.StandardScaler()
x_norm = x_normalize.fit_transform(x)

# Normalize the target variable (optional)
y_normalize = preprocessing.MinMaxScaler()
y_norm = y_normalize.fit_transform(y.reshape(-1, 1))

# Create a Linear Regression model
model = linear_model.LinearRegression()

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = model_selection.train_test_split(x_norm, y_norm.reshape(-1,), test_size=0.1, random_state=40)

# Fit the model to the training data
model.fit(x_train, y_train)

# Make predictions on the test set
y_predict = model.predict(x_test)

# Calculate the mean squared error
error = metrics.mean_squared_error(y_test, y_predict)
print(f"The error using mean squared error = {error}\n")


# Visualize the results
plt.figure(figsize=(14,10))
plt.scatter(y_test, y_predict, marker="o",color='green')
plt.plot(y_predict, color='red',linewidth=2)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Linear Regression Prediction vs. Actual Values")
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.grid(True)
plt.show()

```

Logistic regression (class)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets , linear_model , preprocessing ,metrics,model_selection




x,y=datasets.load_breast_cancer(return_X_y=True)



x_normalize=preprocessing.StandardScaler()

x_norm=x_normalize.fit_transform(x)



model=linear_model.LogisticRegression()
x_train,x_test,y_train,y_test=model_selection.train_test_split(x_norm,y,test_size=0.2,random_state=40,stratify=y)

model.fit(x_train,y_train)

y_predict=model.predict(x_test)

acc=metrics.accuracy_score(y_test,y_predict)

print(f"the accuracy using logistic regression= {acc} \n") 

cm=metrics.confusion_matrix(y_test, y_predict) 

sns.heatmap(cm,annot=True,cmap="Greens")
plt.xlabel("actual values" ,fontdict={"size":10,"color":"green"})
plt.ylabel("Predicted values",fontdict={"size":10,"color":"green"})
plt.legend()
plt.show()




```

Decision tree (class)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets , model_selection ,tree,preprocessing,metrics


x,y=datasets.load_breast_cancer(return_X_y=True)

x_norm=preprocessing.StandardScaler()
x_normalize=x_norm.fit_transform(x)

x_train,x_test,y_train,y_test=model_selection.train_test_split(x_normalize,y,test_size=0.2,random_state=40,stratify=y)

t=tree.DecisionTreeClassifier()
t.fit(x_train,y_train)
y_predict=t.predict(x_test)  
plt.figure(figsize=(10,10),dpi=100)

acc=metrics.accuracy_score(y_predict,y_test)

print(f"the accuracy using trees={acc} \n")
print("\n")
# tree.plot_tree(t) 
# print("\n")


cm=metrics.confusion_matrix(y_test, y_predict) 

sns.heatmap(cm,annot=True,cmap="Greens")
plt.xlabel("actual values" ,fontdict={"size":10,"color":"green"})
plt.ylabel("Predicted values",fontdict={"size":10,"color":"green"})
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import datasets , model_selection ,ensemble,preprocessing,metrics


x,y=datasets.load_breast_cancer(return_X_y=True)
# df=pd.DataFrame(y)
# print(df )

x_train,x_test,y_train,y_test=model_selection.train_test_split(x,y,test_size=0.2,random_state=40,stratify=y)

r=ensemble.RandomForestClassifier(max_depth=10,n_jobs=2,random_state=200)
r.fit(x_train,y_train)
y_predict=r.predict(x_test)  


acc=metrics.accuracy_score(y_predict,y_test)

print(f"the accuracy using random forest={acc} \n")
print("\n") 
cm=metrics.confusion_matrix(y_test,y_predict)

sns.heatmap(cm,annot=True,cmap="Greens")
plt.xlabel("actual values" ,fontdict={"size":10,"color":"green"})
plt.ylabel("Predicted values",fontdict={"size":10,"color":"green"})
plt.legend()
plt.show() 
print("\n") 

sns.scatterplot(x=x_test[:, 0], y=x_test[:,1],hue=y_predict, palette='Greens')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()



```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets , model_selection ,ensemble,preprocessing,metrics,svm


x,y=datasets.load_breast_cancer(return_X_y=True)

x_train,x_test,y_train,y_test=model_selection.train_test_split(x,y,test_size=0.2,random_state=40,stratify=y)

model=svm.SVC(gamma="scale",kernel='linear')
model.fit(x_train,y_train)
y_predict=model.predict(x_test)

acc=metrics.accuracy_score(y_predict,y_test)

print(acc)
print("\n")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
from sklearn import datasets , linear_model , preprocessing ,metrics,model_selection,svm


x,y=datasets.load_iris(return_X_y=True)


x_mean=x.mean()
x_std=x.std()
x_norm=(x-x_mean)/x_std

y_mean=y.mean()
y_std=y.std()
y_norm=(y-y_mean)/y_std


model=svm.SVR(gamma='scale',kernel='linear')
x_train,x_test,y_train,y_test=model_selection.train_test_split(x_norm,y_norm,test_size=0.1,random_state=40)

model.fit(x_train,y_train)

y_predict=model.predict(x_test)

error=metrics.mean_squared_error(y_test,y_predict)

print(error)
```
