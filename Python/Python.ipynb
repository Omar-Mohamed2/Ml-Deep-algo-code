```python
# first built in function
int("102")  
str(103)
float('1.25') 
bool(0)



def myFun(arg1, *argv):
    print("First argument :", arg1)
    for arg in argv:
        print("Next argument through *argv :", arg)
 
 
myFun('Hello', 'Welcome', 'to', 'GeeksforGeeks') 


print("\n")

# must be named

def myFun(**kwargs):
    for key, value in kwargs.items():
        print(f"{key} =={value}")
 
 
# Driver code
myFun(first='Geeks', mid='for', last='Geeks') 


print("\n")


print("example for recursive fun ") 

def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

number = 5
result = factorial(number)
print(f"The factorial of {number} is {result}") 

```

regular expression

```python
import re 

# first apply serach operation

pattren=r"omar"

result=re.search(pattren,"My name is omar mohamed")

print(result.group()) 


# first apply serach operation on one char followed by any number of char expect new line

pattren=r"o...."

result=re.search(pattren,"My name is omar mohamed")

print(result.group()) 



#Character Classes:** Define a set of characters to match.

''' - `[abc]`: Matches any of the characters a, b, or c.
 `[a-z]`: Matches any lowercase letter.
   - `[A-Za-z]`: Matches any letter (both uppercase and lowercase).'''

   #Example: To match any vowel

pattern = r"[A-Z , a-z]"
text = "apple"
result = re.findall(pattern, text )
print(result)  


x_test=r"o+" 
t="hello omar"
result=re.findall(x_test,t)
print(result) 


x_test=r"^hello" 
t="hello omar"
result=re.findall(x_test,t,re.MULTILINE)
print(result) 
 


 # test mail

import re

def is_valid_gmail(email):
    pattern = r'^[a-z,A-Z,0-9,_%+-]+@gmail+.com$'
    re.match(pattern, email) 
    if re.match:
        print("found")
    else:
        print("not")

# Test cases
print(is_valid_gmail("Example@gmail.com"))  # Should return True
print(is_valid_gmail("invalid@example.com"))  # Should return False

import re

def is_valid_phone(phone):
    pattern = r'^\+20[1][0125][0-9]{8}$'
    return re.match(pattern, phone)

# Test cases
print(is_valid_phone("0482293115"))  # Should return True
print(is_valid_phone("+201029882389"))  # Should return True 


```

```python
# num=int(input("please enter the number"))

# try:
#    1/num
# except ZeroDivisionError:
#    print("value error")
# else:
#    pass  



class MyCustomError(Exception):
    pass

def my_function(value):
    if value == 0:
        raise MyCustomError("Value cannot be zero")

try:
    my_function(0)
except MyCustomError as e:
    print("Custom error:", e) 


def divide(x,y):
    try:
      if y==0:
        raise ZeroDivisionError("can not divide on zero") 
    except ZeroDivisionError as z:
        print(z)
    else:
        print(x/y)
divide(2,0)
```
